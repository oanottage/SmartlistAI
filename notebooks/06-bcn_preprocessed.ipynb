## PREPROCESSING FILE FOR KNMEANS CLUSTERING ALGORITHM

```python
df_knm_businesses = df_business_districts.merge(df_final, left_on="Area", right_index=True, how = "left")
```

```python
#Businesses with websites
df_knm_businesses['has_website'] = np.where(df_knm_businesses['website'].notna(), 1, 0)
df_knm_businesses = df_knm_businesses[df_knm_businesses['permanently_closed'].isna()]
```

```python
#Remove unncessary columns
df_knm_businesses.drop(columns=['category','formatted_phone_number',
       'international_phone_number','name', 'opening_hours', 'place_id','types', 'url', 'user_ratings_total','Count','website','permanently_closed','price_level','business_status'],inplace=True)
```

```python
#One hot encode the areas
df_knm_businesses = pd.get_dummies(df_knm_businesses, columns=['Area'], drop_first=True)
```

```python
median_rating = df_knm_businesses['rating'].median()
df_knm_businesses['rating'].fillna(median_rating, inplace=True)
```
